### Report ###

1. Testing the trained model with an Epoch of 5 wasn't able to produce a Test Accuracy of over 70%.
2. Increasing the Epoch to 10 seemed to work to get the accuracy over 70%.
3. For the Optimizer, trying Adam with the same lr=0.01 value didn't seem to produce very accurate training.
4. Looking up the best lr values for Adam revealed that it should be lr=0.001 as the default.
5. I'm not sure if it was just the specific run or if the optimization is different, but the Adam optimizer seemed to have gotten slightly better results with a lower epoch.

#### Adam ####
##### 10 epochs - lr=0.01 :: fc1 = nn.Linear(1024, 512, True) :: fc2 = nn.Linear(512, 10, True) #####
- Accuracy: 10%
- Time taken for training: ~3 min
- What helps, what does not: As explained above, the lr value of 0.01 for this optimizer does not work at all. Please see the next block for lr=0.001.

#### Adam ####
##### 10 epochs - lr=0.001 :: fc1 = nn.Linear(1024, 512, True) :: fc2 = nn.Linear(512, 10, True) #####
- Accuracy: 72%
- Time taken for training: ~3 min
- What helps, what does not: Changing the lr value to 0.001 for this worked a lot better!

#### Adam ####
##### 10 epochs :: lr=0.001 fc1 = nn.Linear(1024, 1024, True) :: fc2 = nn.Linear(1024, 10, True) #####
- Accuracy: 73%
- Time taken for training: ~3.25 min
- What helps, what does not: Changing the output paramaters for fc1 and input for fc2 seemed to increase the accuracy a little bit but with sacraficing more a little bit more time for training.

#### SGD ####
##### 10 epochs - lr=0.01 :: fc1 = nn.Linear(1024, 512, True) :: fc2 = nn.Linear(512, 10, True) #####
- Accuracy: 68%
- Time taken for training: ~2.5 min
- What helps, what does not: Changing from the default 5 epochs helped with the accuracy a lot, but still below the requested 70%. Going to try upping the epochs to 15.

#### SGD ####
##### 15 epochs - lr=0.01 :: fc1 = nn.Linear(1024, 512, True) :: fc2 = nn.Linear(512, 10, True) #####
- Accuracy: 72%
- Time taken for training: ~4 min
- What helps, what does not: Increasing the epochs seemed to get the training accuracy to the level of the Adam optimizer at 10 epochs. While Adam seems to typically go a little slower given the same number of epochs, it seems worth it in the end to have a higher accuracy and overall faster training time.
